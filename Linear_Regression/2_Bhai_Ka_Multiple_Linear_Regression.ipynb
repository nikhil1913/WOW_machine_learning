```python
import numpy as np
import pandas as pd
```

```python
#  we will be using OLS method to train our model
class Bhai_ka_Multpile_LinearRegression():
    
    def __init__(self,optimizer = 'OLS',epoch = 500,lr = 0.01):
        self.optimizer = optimizer
        self.lr = lr
        self.epoch = epoch
        self.coeff = None
        self.intercept = None

    def initilize_params(self,X):
        n_features = X.shape[1]
        self.coeff = np.zeros(n_features)   # initilizing initital weights with 0
        self.intercept = 0
        
    #training happens using ols method
    def ols(self,X,y):
        X = np.insert(X,0,1,axis =1 )
        betas = np.linalg.inv(np.dot(X.T,X)).dot(X.T).dot(y)
            
        self.intercept = betas[0]
        self.coeff = betas[1:]

    #here training happens using SGD
    def SGD(self,X,y):
        # here first we need to initilize the params
        self.initilize_params(X)

        for i in range(self.epoch):
            # in stochastic gradient descent out batch size = 1
            for j in range(X.shape[0]):
                idx = np.random.randint(0,X.shape[0])
                Xi = X[idx]
                yi = y[idx]

                yhat = self.intercept + np.dot(Xi,self.coeff)

                # doing derivative 
                Dcoeff = -2 * (yi - yhat) * Xi
                Dintercept = -2 * (yi - yhat)

                # updating the parameter
                self.coeff = self.coeff - self.lr * Dcoeff
                self.intercept = self.intercept - self.lr * Dintercept
                
            if(i % 100 == 0) :
                y_pred = self.intercept + np.dot(X,self.coeff)
                print(f'after epoch {i} loss is { np.mean((y-y_pred)**2) }')
        
    def fit(self,X,y):
        if self.optimizer == 'OLS' :
            self.ols(X,y)
        if self.optimizer == 'SGD' :
            self.SGD(X,y)
        
    def predict(self,X):
        y_pred = self.intercept + np.dot(X,self.coeff)
        return y_pred
```

```python
from sklearn.datasets import make_regression
X, y = make_regression(n_samples=1000, n_features=3, noise=10, random_state=42)
```

```python
y[0:5]
```

```python
from sklearn.model_selection import train_test_split

X_train , X_test , y_train , y_test = train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
X_train.shape , y_train.shape
```

```python
X_test.shape , y_test.shape
```

```python
from sklearn.linear_model import LinearRegression
sklearn_model = LinearRegression()
sklearn_model
```

```python
sklearn_model.fit(X_train, y_train)
```

```python
from sklearn.metrics import mean_absolute_error , mean_squared_error , r2_score

y_pred_sklearn = sklearn_model.predict(X_test)

print(f'sklearn ki test score is {sklearn_model.score(X_test,y_test)*100:.2f} %')
print(f'sklearn ki test MSE is {mean_squared_error(y_test,y_pred_sklearn)} ')
print(f'sklearn ki test MAE is {mean_absolute_error(y_test,y_pred_sklearn)} ')
print(f'sklearn ki test R2_score is {r2_score(y_test,y_pred_sklearn):.3f} %')
```
