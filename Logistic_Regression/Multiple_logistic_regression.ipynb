```python
import numpy as np

class Bhai_Ka_Logistic_Regression_model:

    def __init__(self,epoch = 1000,lr = 0.05):
        self.epoch = epoch
        self.weights = None
        self.lr = lr
        self.intercept = None
        self.coeff = None
        
    def weight_initilization(self,X,y):
        self.weights = np.ones((X.shape[1], y.shape[1]))

    def one_hot_en(self, y, num_classes):
        
        one_hot = np.zeros((y.size, num_classes))
        one_hot[np.arange(y.size), y] = 1
        return one_hot
        
    def soft_max(self, Z):
        m, k = Z.shape # k is num_classes
        output = np.zeros((m, k))

        for i in range(m):
            that_row = Z[i, :]
            max_val = np.max(that_row)
            row_stable = that_row - max_val 

            exp_values = np.exp(row_stable)
            exp_sum = np.sum(exp_values)
            
            output[i, :] = exp_values / exp_sum
            
        return output
        
    def fit(self,X,y):
        m, n = X.shape  
        num_classes = len(np.unique(y))
        X_new = np.insert(X,0,1,axis=1)
        
        y_enc = self.one_hot_en(y,num_classes)
        
        self.weight_initilization(X_new,y_enc)
        
        
        for i in range(self.epoch):
            Z = np.dot(X_new,self.weights)
            y_hat = self.soft_max(Z)
            self.weights = self.weights + self.lr*(np.dot(X_new.T,(y_enc-y_hat)))/m
            
        self.intercept = self.weights[0]
        self.coeff = self.weights[1:]

    def predict(self, X):
        X_new = np.insert(X, 0, 1, axis=1)
        Z = np.dot(X_new, self.weights)
        y_pred = self.soft_max(Z)
        
        return np.argmax(y_pred, axis=1)
```

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split

# Data load karo
data = load_iris()
X_train, X_test, y_train, y_test = train_test_split(data.data, data.target, test_size=0.2)

# Bhai ka bhai_ka_model initialize karo
bhai_ka_model = Bhai_Ka_Logistic_Regression_model(epoch=5000, lr=0.1)
bhai_ka_model.fit(X_train, y_train)

# Predictions nikal lo
preds = bhai_ka_model.predict(X_test)
print(f"Bhai ki accuracy: {np.mean(preds == y_test) * 100}%")

print(f'bhai ka model ka intercept {bhai_ka_model.intercept}')
print(f"bhai ka model k coeff{ bhai_ka_model.coeff}")
```
